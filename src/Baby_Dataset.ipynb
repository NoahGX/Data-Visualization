# **Data Visualization on Baby Dataset**

This Jupyter Notebook provides a comprehensive analysis of a dataset containing various attributes of babies and their mothers. The analysis includes visualizations of data distributions, comparative studies between different groups, and exploration of correlations between various factors. The goal is to find insights about factors that affect birth weights and maternal health.

***

## **1. Setup and Data Loading**

This section provides an initial look at the data and prepares for further detailed analysis.

```python
# Importing Necessary Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objs as go
import plotly.figure_factory as ff
import plotly.offline as py
import plotly.io as pio
```

```python
# Configuration for Plotly to work with Jupyter Notebook
py.init_notebook_mode(connected=True)
```

```python
# Data Loading
baby = pd.read_csv("../data/baby.csv")
print(baby.head())
```

***

## **2. Visualizing Data Distribution**

Exploring different visualization methods to understand the data distribution.

### *2.1 ~ Bar Plot Analysis*

```python
# Count of maternal smokers and non-smokers
smoker_counts = baby["Maternal Smoker"].value_counts()
print(smoker_counts)
```

```python
# Plot using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(smoker_counts.index, smoker_counts.values, color='skyblue')
plt.title('Distribution of Maternal Smokers')
plt.xlabel('Maternal Smoker')
plt.ylabel('Count')
plt.show()
```

```python
# Plot using Seaborn
plt.figure(figsize=(10, 6))
sns.countplot(data=baby, x="Maternal Smoker", hue="Maternal Smoker", palette='viridis')
plt.title('Distribution of Maternal Smokers using Seaborn')
plt.show()
```

```python
# Plot using Plotly
fig = px.histogram(baby, x="Maternal Smoker", color="Maternal Smoker", title="Interactive Distribution of Maternal Smokers")
fig.show()
```

### *2.2 ~  Histogram Analysis*

```python
# Creating histograms and rug plots for Maternal Pregnancy Weight
sns.histplot(data=baby, x="Maternal Pregnancy Weight", kde=True, color='blue', bins=10)
sns.rugplot(data=baby, x="Maternal Pregnancy Weight", color="red")
plt.title('Histogram and Rugplot of Maternal Pregnancy Weight')
plt.show()
```

```python
# Interactive histogram using Plotly
fig = px.histogram(baby, x="Maternal Pregnancy Weight", nbins=10, title="Interactive Histogram of Maternal Pregnancy Weight")
fig.show()
```

### *2.3 ~ Box and Violin Plots*

```python
# Box plot to compare the distribution of birth weights
plt.figure(figsize=(12, 6))
sns.boxplot(data=baby, y="Birth Weight")
plt.title('Box Plot of Birth Weight')
plt.show()
```

```python
# Violin plot to compare the distribution of birth weights
plt.figure(figsize=(12, 6))
sns.violinplot(data=baby, y="Birth Weight")
plt.title('Violin Plot of Birth Weight')
plt.show()
```

### *2.4 ~ Comparative Analysis of Distributions*

```python
# Comparing the birth weights between babies of smokers and non-smokers
plt.figure(figsize=(12, 6))
sns.histplot(data=baby, x="Birth Weight", hue="Maternal Smoker", element='step', stat='density', common_norm=False)
plt.title('Density Plot of Birth Weight by Smoking Status')
plt.show()
```

```python
# Comparative Box and Violin Plots
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
sns.boxplot(data=baby, x="Maternal Smoker", y="Birth Weight", ax=axes[0])
sns.violinplot(data=baby, x="Maternal Smoker", y="Birth Weight", ax=axes[1])
axes[0].set_title('Box Plot of Birth Weight by Smoking Status')
axes[1].set_title('Violin Plot of Birth Weight by Smoking Status')
plt.show()
```

### *2.5 ~ Scatter Plot Analysis*

```python
# Investigating relationships between maternal attributes and birth weight
plt.figure(figsize=(10, 6))
sns.scatterplot(data=baby, x="Maternal Height", y="Birth Weight", hue="Maternal Smoker")
plt.title('Scatter Plot of Birth Weight vs. Maternal Height')
plt.xlabel('Maternal Height')
plt.ylabel('Birth Weight')
plt.show()
```
