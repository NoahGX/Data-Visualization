# **Heatmap Visualization using Matplotlib and Seaborn**

This Jupyter Notebook demonstrates how to create heatmaps using both Matplotlib and Seaborn for visualizing quantitative data. Heatmaps provide an effective way to identify patterns in data, and we will explore both diverging and sequential color maps in our examples.

***

## **1. Create the DataFrame**

Here we define a small dataset with numerical values and create a DataFrame for our heatmap visualizations.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define index and columns for the DataFrame
index_labels = ['1', '2', '3', '4']
column_labels = ['A', 'B', 'C', 'D']

# Create a DataFrame with predefined values
data = [
    [10, 20, 30, 40],
    [50, 30, 8, 15],
    [25, 14, 41, 8],
    [7, 14, 21, 28]
]
df = pd.DataFrame(data, columns=column_labels, index=index_labels)

# Display the DataFrame for reference
df
```

***

## **2. Heatmap using Matplotlib**

In this section, we will generate a heatmap using **Matplotlib** with a diverging color map (`RdYlBu`) to visualize quantitative data.

```python
# Heatmap using Matplotlib for quantitative data visualization
plt.figure(figsize=(6, 4))
plt.imshow(df, cmap="RdYlBu", aspect='auto')
plt.colorbar(label='Value Range')

# Customize axis ticks and labels
plt.xticks(ticks=range(len(df.columns)), labels=df.columns)
plt.yticks(ticks=range(len(df.index)), labels=df.index)

# Add title and labels
plt.title('Heatmap using Matplotlib', fontsize=14)
plt.xlabel('Columns')
plt.ylabel('Index')
plt.tight_layout()

# Display the Matplotlib heatmap
plt.show()
```

***

## **3. Heatmap using Seaborn with Diverging Colormap**

Next, we generate a heatmap using **Seaborn** with a diverging color map (`RdYlGn`). 

Seaborn offers enhanced visualization options, including the ability to annotate values directly on the heatmap.

```python
# Heatmap using Seaborn with diverging colormap 'RdYlGn'
plt.figure(figsize=(6, 4))
sns.heatmap(df, cmap='RdYlGn', annot=True, fmt='d')
plt.title('Heatmap using Seaborn (RdYlGn)', fontsize=14)
plt.tight_layout()

# Display the Seaborn heatmap
plt.show()
```

## **4. Heatmap using Seaborn with Sequential Colormap (Viridis)**

Finally, we explore a heatmap using Seaborn with a **sequential colormap** (`viridis`), which is often useful for showing progressive or cumulative data.

```python
# Heatmap using Seaborn with sequential colormap 'viridis'
plt.figure(figsize=(6, 4))
sns.heatmap(df, cmap='viridis', annot=True, fmt='d')
plt.title('Heatmap using Seaborn (Viridis)', fontsize=14)
plt.tight_layout()

# Display the Seaborn heatmap
plt.show()
```
